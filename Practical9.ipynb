```python
import pandas as pd
import numpy as np
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("titanic")
```

```python
df.info()
```

```python
df = df.dropna(subset=['age'])
```

```python
sns.set_theme(style="whitegrid")
```

```python
plt.figure(figsize=(10,6))

sns.boxplot(
    x='sex',
    y='age',
    hue='survived',
    data=df
)
```

```python
plt.legend(title="Survived (0 = No, 1 = Yes)")
```

```python
plt.legend(handles, ['Not Survived', 'Survived'], title="Survival Status")

plt.title("Distribution of Age by Gender and Survival Status", fontsize=14)
plt.xlabel("Gender", fontsize=12)
plt.ylabel("Age", fontsize=12)
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df['age'], bins=30, kde=True)

plt.title("Histogram of Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(6,5))

sns.countplot(x='survived', data=df)

plt.title("Count of Survival (0 = No, 1 = Yes)")
plt.show()
```

```python
plt.figure(figsize=(6,5))

sns.barplot(x='sex', y='age', data=df)

plt.title("Average Age by Gender")
plt.show()
```

```python
import pandas as pd


df_corr = df[['age', 'fare', 'survived', 'pclass']]

plt.figure(figsize=(8,6))

sns.heatmap(df_corr.corr(), annot=True, cmap='coolwarm')

plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.pairplot(df[['age', 'fare', 'survived', 'pclass']], hue='survived')

plt.show()
```
